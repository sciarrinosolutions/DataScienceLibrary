## Importing with a Datetime Column

```python
df = pd.read_csv(filename, parse_dates=["Date"])  # Throws a UserWarning since Date is not in a common datetime format
df.dtypes
```

```python
df = pd.read_csv(filename, parse_dates=["Date_Clean"]) # Date_Clean is already in a common datetime format
df.dtypes
```

## Formatting a Datetime Column

```python
# Convert the Date_Clean column to datetime format and assign it to a new column called Date
df['Date'] = pd.to_datetime(df['Date_Clean'], format='%b %d %Y')
df.dtypes
```

```python
df.drop(columns=['Date_Clean'], inplace=True)
df.head(5)
```

## Extracting Datetime Values

```python
# Retrieve the month, day, year, day of week, quarter, and weekday number from the Date column
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Year'] = df['Date'].dt.year
df['DayOfWeek'] = df['Date'].dt.day_name()
df['Quarter'] = df['Date'].dt.quarter
df['WeekdayNumber'] = df['Date'].dt.weekday
```

```python
df.head(10)
```

## Grouped Chronological Sortation

```python
# Chronologically sort the data by FeatureName and Date, then reset the index
featureName = 'Team'
df = df.sort_values([featureName, 'Date']).reset_index(drop=True)
df.tail(10)
```

## Filtering Before or After Specific Dates

```python
# Filtering After a Certain Date
df_post_april = df[df['Date'] >= '2018-05-01'].reset_index(drop=True)
df_post_april.head(10)
```

```python
# Filtering Before a Certain Date
df_pre_april = df[df['Date'] < '2018-04-01'].reset_index(drop=True)
df_pre_april.head(10)
```
